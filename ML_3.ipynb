```python
import matplotlib.pyplot as plt
x=2 #starting point
lr=0.01 # learning rate= step size
prev_step_size=1
precision=0.000001
max_iteration=10000 #for how long while loop should execute
iters=0#initial vale for iteration
gd=[]
gf=lambda x : (x+3)**2 #gradient descent function
while((precision<prev_step_size) and (iters<max_iteration)):
    prev=x
    x=x-lr*gf(prev)
    prev_step_size=abs(x-prev)
    iters+=1
    print("Iteration :",iters)
    print("Values :",x)
    gd.append(x)
print("Local Minima :",x)
plt.plot(gd)
plt.show()
```
